```python
import pandas as pd
from sqlalchemy import create_engine
#from secret import connection_string
```

```python
csv_file = "../Run Data/world-marathon-majors/Female_Elite_Boston.csv"
female_runner_data = pd.read_csv(csv_file,low_memory= True, encoding ='iso8859_15')
female_runner_data.head(20)
```

```python
female_runner_data = female_runner_data.dropna()
clean_female_runner_data = female_runner_data[['Year','Athlete', 'Country','Time']].copy()
```

```python

clean_female_runner_data.set_index("Year",inplace = True)
```

```python
clean_female_runner_data.head()
```

```python
csv_file = "../Run Data/world-marathon-majors/Male_Elite_Boston.csv"

male_runner_data = pd.read_csv(csv_file,low_memory= True, encoding ='iso8859_15')
male_runner_data.head(20)
```

```python
male_runner_data = male_runner_data.dropna()
clean_male_runner_data = male_runner_data[['Year','Athlete', 'Country','Time']].copy()

```

```python
clean_male_runner_data.set_index("Year",inplace = True)
```

```python
clean_male_runner_data
```

```python
clean_female_runner_data.to_sql(name ='female_boston', con = engine, if_exists = 'replace')
```
